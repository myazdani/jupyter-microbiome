## Only keep the species and subject columns and transpose the data:

```r
df.rel = df[,c(9:ncol(df))]
names(df.rel)
```

```r
## Tranpose data
# first remember the names
row.names(df.rel) = df.rel$species
df.rel = df.rel[,-1]
n <- row.names(df.rel)
# transpose all but the first column (name)
species.df <- as.data.frame(t(df.rel), stringsAsFactors = FALSE)
colnames(species.df) <- n
head(species.df)
```

## Apply and visualize PCA

```r
pca.result = prcomp(species.df, scale = FALSE)
```

```r
pca.df = as.data.frame(pca.result$x)
pca.df$patient = row.names(pca.df)
pca.df$patient.type = substr(pca.df$patient, 1, 2)
```

## Reshape data and visualize

```r
library(reshape2)
```

```r
species.df$subject = row.names(species.df)
```

```r
species.m = melt(species.df, id.vars = "subject")
```

```r
head(species.m)
```

```r
names(species.m)[2:3] = c("species", "species.abundance")
```

```r
microbe.keys = df[,c(3:9)]
head(microbe.keys)
```

```r
micorbe.df = merge(species.m, microbe.keys)
```

```r
head(micorbe.df )
```

```r
LS = subset(micorbe.df, grepl("LS", micorbe.df$subject))
head(LS)
```

```r
ggplot(LS, aes(x = subject, y = species.abundance, fill = phylum, order = phylum)) + 
            geom_bar(stat="identity") + xlab("") + guides(fill=FALSE) -> p.LS
plotly_POST(p.LS, filename = "LS-bar-graph-phylum", world_readable=TRUE)
```
