# PCA of Species Relative Abundance 

```r
setwd("~/MICROBIOME/")
```

```r
df = read.csv("./processedData/bin-2013-1015-bae-ref_cov_abundance-clean/ref_species_cov_abundance.txt-Table 1.csv", header = TRUE, stringsAsFactors = FALSE)
head(df)
```

## Only keep the species and subject columns and transpose the data:

```r
df.rel = df[,c(9:ncol(df))]
names(df.rel)
```

```r
## Tranpose data
# first remember the names
row.names(df.rel) = df.rel$species
df.rel = df.rel[,-1]
n <- row.names(df.rel)
# transpose all but the first column (name)
species.df <- as.data.frame(t(df.rel), stringsAsFactors = FALSE)
colnames(species.df) <- n
head(species.df)
```

## Apply and visualize PCA

```r
pca.result = prcomp(log10(.00000001+ species.df), scale = FALSE)
```

```r
pca.df = as.data.frame(pca.result$x)
pca.df$patient = row.names(pca.df)
pca.df$patient.type = substr(pca.df$patient, 1, 2)
```
